## Q3 Find the Origin airport which had the largest departure delay in the month of January

```python
flights_in_jan=ftdata[ftdata['MONTH']==1]

max_delay_flight=flights_in_jan[flights_in_jan['DEPARTURE_DELAY'] >= flights_in_jan['DEPARTURE_DELAY'].max()]

print(max_delay_flight['ORIGIN_AIRPORT'])
```

## Q4 Find out which carrier has the largest delay on Weekends.

```python
df4 = ftdata[ftdata['DAY_OF_WEEK'].isin([6,7])]

df4_1 = df4.nlargest(1,['DEPARTURE_DELAY'])

df4_1[['DEPARTURE_DELAY', 'AIRLINE','FLIGHT_NUMBER']]

```

## Q5 Which airport has the most cancellation of flights?

```python
df5 = ftdata[ftdata['CANCELLED']==1]

destinations = df5['DESTINATION_AIRPORT'].value_counts()

origin = df5['ORIGIN_AIRPORT'].value_counts()


print('destination airport', destinations.nlargest(1))
print('origin airport', origin.nlargest(1))
```

## Q6 Find the percent of flights cancelled for each carrier.

```python
df6 = ftdata.groupby('AIRLINE').mean()['CANCELLED']
print(df6)
df6.nlargest(1)
```

## Q7 Find the largest departure delay for each carrier

```python
df7 = ftdata.groupby('AIRLINE').max()['DEPARTURE_DELAY']
print(df7)
df7.nlargest(1)
```

## Q8 Find the largest departure delay for each carrier for each month

```python
for i in range(1,12):
    df8 = ftdata[ftdata['MONTH'] == i]
    df8_1 = df8.groupby('AIRLINE').max()['DEPARTURE_DELAY']
    print('Month',i,df8_1)
```

## Q9 For each carrier find the average Departure delay

## Q11 Which date of year has the highest rate of flight cancellations? Rate of flight cancellation is calculated by deviding number of canceled flights by total number of flights.

```python
df11 = ftdata.groupby(['MONTH','DAY']).mean()['CANCELLED']
print(df11)
df11.nlargest(1)
```

## Q12 Calculate the number of flights to each destination state For each carrier, for which state do they have the largest average delay?

```python
df12 = ftdata['AIRLINE'].unique()
l12 = list(df12)
for i in l12:
    df12_1 = ftdata[ftdata['AIRLINE'] == i]
    df12_2 = df12_1.groupby('DESTINATION_AIRPORT').count()['FLIGHT_NUMBER']
    print(i,df12_2)
```

```python
for i in l12:
    df12_1 = ftdata[ftdata['AIRLINE'] == i]
    df12_3 = df12_1.groupby('DESTINATION_AIRPORT').mean()['CANCELLED']
    print(i, df12_3.nlargest(1))
```
